### By DP 

* It starts with an empty matrix that has the size of the length of the strings. 
* Both the first row and column, starting from zero, are indexed increasingly
* Next, two loops follow to compare the strings letter by letter - row-wise, and column-wise
* If two letters are equal, the new value at position [x, y] is the minimum between the value of position [x-1, y] + 1, position [x-1, y-1], and position [x, y-1] + 1.
* Otherwise, it is the minimum between the value of position [x-1, y] + 1, position [x-1, y-1] + 1, and position [x, y-1] + 1

```python
def levenstein(s1,s2):
    
    size1 = len(s1)+1
    size2 = len(s2)+1
        
    matrix = np.zeros((size1,size2))
    
    for x in range(size1):
        matrix [x, 0] = x
        
    for y in range(size2):
        matrix [0 ,y] = y
        
    for x in range(1,size1):
        for y in range(1,size2):
            if(s1[x-1]==s2[y-1]):
                matrix[x,y]=min(matrix[x-1,y]+1, matrix[x-1,y-1], matrix[x,y-1]+1)
                
            else:
                matrix[x,y]=min(matrix[x-1,y]+1, matrix[x-1,y-1]+1, matrix[x, y-1]+1)
        
    print(matrix)
    return matrix[size1-1,size2-1]
```

```python
levenstein("happier","happy")
```

## Jaccard's Cofficient 

Jaccard cofficient compares members for two sets to see which members are shared and which are distinct. <br>Itâ€™s a measure of similarity for two sets of data, with range from 0% to 100%.<br> Higher percentage, more similar two populations

```python
list1=(['my','name','is','pawan','jain'])
list2=(['my','sisters','name','is','Ekta','jain'])
```

```python
jaccard_distance(list1,list2)
```

### Code

* Count the number of members which are shared between both sets.
* Count the total number of members in both sets (shared and un-shared).
* Divide the number of shared members (1) by the total number of members (2).
* Multiply the number you found in (3) by 100.

```python
happy = 'my name is pawan jain'
unhappy = 'my sisters name is ekta jain'
```

```python
def jacc(s1,s2):
    s1,s2 = s1.split(),s2.split()
    inte = set(s1).intersection(set(s2))
    un = set(s1).union(set(s2))
    print(inte)
    print(un)
    
    dist = 1-len(inte)/len(un)
    sim = len(inte)/len(un)
    
    return (dist,sim)
```

```python
jacc(happy,unhappy)
```

## Binary Similarity Measure 

```python
def bina(a1,a2):
    if (a1==a2):
        r=0.0
    else:
        r=1.0

    return r
```

```python
bina('happy','unhappy')
```

## Masi Distance
Distance metric that takes into account partial agreement when multiple labels are assigned

```python
def masi(l1,l2):
    
    l1,l2 = set(l1), set(l2)
    
    len_int, len_un = len(l1.intersection(l2)), len(l1.union(l2))
    len_l1, len_l2 = len(l1), len(l2)
    
    if len_l1==len_l2 and len_l1==len_int:
        m=1.0
    elif len_int == min(len_l1,len_l2):
        m=0.67
    elif len_int>0:
        m=0.33
    else:
        m=0.0
        
    return 1-(len_int/len_un)*m
```

```python
l1 = [10,20,30,40]
l2 = [30,50,70]
```

```python
print(masi(l1,l2))
```
